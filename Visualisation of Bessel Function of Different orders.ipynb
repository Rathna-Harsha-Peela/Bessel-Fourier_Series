```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv

def plot_bessel_function(n, x_range):

    #Plots the Bessel function of order n over the specified x range.
    # n is the order of the Bessel Function
    # x_start and x_end are the start and end values of interval in which Bessel function is simulated
    
    x = np.linspace(x_range[0], x_range[1], 400)
    Jn_values = jv(n, x)  # Compute the Bessel function of order n

    plt.figure(figsize=(10, 5))
    plt.plot(x, Jn_values, label=f'$J_{n}(x)$', color='blue')
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.title(f'Bessel Function of Order {n}: $J_{n}(x)$')
    plt.xlabel('$x$')
    plt.ylabel(f'$J_{n}(x)$')
    plt.grid()
    plt.legend()
    plt.show()

def main():
    # User input for order of Bessel function
    n = int(input("Enter the order of the Bessel function (n): "))
    x_start = float(input("Enter the start of the x range: "))
    x_end = float(input("Enter the end of the x range: "))
    
    plot_bessel_function(n, (x_start, x_end))

main()
```
